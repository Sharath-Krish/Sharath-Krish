# Data Cleaning and Manupilation

```python
df.info()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(df.isnull(),
            cmap='plasma',
            yticklabels=False,
            cbar=False)
plt.title('Missing values',fontsize=15)
plt.xticks(fontsize=10)
plt.show()
```

```python
df['retail_price'].fillna(df['retail_price'].median(),inplace=True)
df['discounted_price'].fillna(df['discounted_price'].median(),inplace=True)
```

```python
#Check for duplicate values
df.duplicated().value_counts()
```

```python
#Coverting timestamp column into datetime 

df['crawl_timestamp'] = pd.to_datetime(df['crawl_timestamp'])
```

```python
df['crawl_year'] = df['crawl_timestamp'].apply(lambda x:x.year)
df['crawl_month'] = df['crawl_timestamp'].apply(lambda x:x.month)
```

```python
#Dividing category column to multiple sub categories
print(df.product_category_tree[1],'\n')
for i in df.product_category_tree[10].split('>>'):
    print(i)
```

```python
# Here we are creating Main category column from category tree 
df['Main_category'] = df['product_category_tree'].apply(lambda x:x.split('>>')[0][2:].strip('[]"').strip())
```

```python
# Here we are fetching secondary,tertiary,quaternary categories from the category tree. Also using try except statements to check for errorsabs
def secondary(x):
    try:
        return x.split('>>')[1][1:].strip('[]"').strip()
    except IndexError:
        return 'None'

def tertiary(x):
    try:
        return x.split('>>')[2][1:].strip('[]"').strip()
    except IndexError:
        return 'None'
    
def quaternary(x):
    try:
        return x.split('>>')[3][1:].strip('[]"').strip()
    except IndexError:
        return 'None'
```

```python
df['Secondary_category'] = df['product_category_tree'].apply(secondary)
df['Tertiary_category'] = df['product_category_tree'].apply(tertiary)
df['Quaternary_category'] = df['product_category_tree'].apply(quaternary)
```

# Discount Sale

```python
# Create a discount percentage column
# Discount Percentage = ((retail price-discount price)/retail price)*100
```

```python
df['discount_%']=round(((df['retail_price']-df['discounted_price'])/(df['retail_price']))*100,0)
```

```python
# We shall create new dataframes consisting category wise products ,discount percentage mean and count of each product
```

```python
Main_cat_disc = pd.DataFrame(df.groupby('Main_category').agg({'discount_%':[np.mean],'Main_category':['count']}))
sec_cat_disc = pd.DataFrame(df.groupby('Secondary_category').agg({'discount_%':[np.mean],'Secondary_category':['count']}))
ter_cat_disc = pd.DataFrame(df.groupby('Tertiary_category').agg({'discount_%':[np.mean],'Tertiary_category':['count']}))
Qua_cat_disc = pd.DataFrame(df.groupby('Quaternary_category').agg({'discount_%':[np.mean],'Quaternary_category':['count']}))
```

```python
Main_cat_disc.columns = ['_'.join(col) for col in Main_cat_disc.columns]
sec_cat_disc.columns = ['_'.join(col) for col in sec_cat_disc.columns]
ter_cat_disc.columns = ['_'.join(col) for col in ter_cat_disc.columns]
Qua_cat_disc.columns = ['_'.join(col) for col in Qua_cat_disc.columns]
```

```python
Main_cat_disc.head()
```

```python
Main_cat_disc = Main_cat_disc.sort_values(by=['Main_category_count'],ascending=False)[:10]
sec_cat_disc = sec_cat_disc.sort_values(by=['Secondary_category_count'],ascending=False)[:10]
ter_cat_disc = ter_cat_disc.sort_values(by=['Tertiary_category_count'],ascending=False)[:10]
Qua_cat_disc = Qua_cat_disc.sort_values(by=['Quaternary_category_count'],ascending=False)[:10]
```
